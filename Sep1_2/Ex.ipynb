```python
def compute_cost(x,y,w):
  f_wb = w[0]*x+w[1]
  cost = np.sum((f_wb-y)**2)
  cost/=(2*len(x))
  return cost
```

```python
def compute_gradient(x,y,w):
  dw = np.zeros(len(w))
  fw = 0
  for i in range(len(w)):
    fw += w[i]*x**i

  for j in range(len(w)):
    dw[j]=np.sum((fw-y)*x**j)/len(x)
    return dw
```
